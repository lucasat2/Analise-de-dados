```python
CREATE TABLE `navios`.`tb_navios` (
  `nome_navio` VARCHAR(50) NULL,
  `mes_ano` VARCHAR(10) NULL,
  `classificacao_risco` VARCHAR(15) NULL,
  `indice_conformidade` VARCHAR(15) NULL,
  `pontuacao_risco` INT NULL,
  `temporada` VARCHAR(200) NULL);
```

```python
2- Carregar o csv na tabela em data import wizard 
```

# SELECT

```python
SELECT nome_navio, mes_ano FROM navios.tb_navios;
```

# WHERE

```python
SELECT DISTINCT nome_navio, classificacao_risco, temporada 
FROM tb_navios 
WHERE classificacao_risco = 'D';
```

# ORDER BY

```python
SELECT DISTINCT nome_navio, classificacao_risco, temporada 
FROM tb_navios 
WHERE classificacao_risco = 'D'
ORDER BY nome_navio;
```

# OPERADORES LOGICOS

## AND

```python
SELECT DISTINCT nome_navio, classificacao_risco, pontuacao_risco, temporada 
FROM tb_navios 
WHERE classificacao_risco = 'D' AND pontuacao_risco > 1000
ORDER BY nome_navio;
```

## OR

```python
SELECT DISTINCT nome_navio, classificacao_risco, pontuacao_risco, temporada 
FROM tb_navios 
WHERE classificacao_risco = 'D' OR pontuacao_risco > 3000
ORDER BY nome_navio;
```

## IN

```python
SELECT DISTINCT nome_navio, classificacao_risco, indice_conformidade, temporada 
FROM tb_navios 
WHERE classificacao_risco IN ('A','B') AND indice_conformidade > 90
ORDER BY nome_navio;
```

## limitar o numero de linhas (nao use como filtro)

```python
SELECT DISTINCT nome_navio, classificacao_risco, indice_conformidade, temporada 
FROM tb_navios 
WHERE classificacao_risco IN ('A','B') AND indice_conformidade > 90
ORDER BY indice_conformidade
LIMIT 10;
```

## EM abril de 2018 alguma embarcação teve indice de conformidade 100% e pontuacao de risco igual a 0?

```python
SELECT DISTINCT nome_navio, classificacao_risco, indice_conformidade, temporada 
FROM tb_navios 
WHERE classificacao_risco IN ('A','B') AND indice_conformidade > 90
ORDER BY nome_navio;
```

## Quais embarcações possuem pontuação de risco igual a 310?

```python
SELECT  nome_navio, pontuacao_risco 
FROM tb_navios 
WHERE pontuacao_risco = 310 
```

## Quais embarcações têm classificação de risco A e índice de conformidade maior ou igual a 95%?

```python
SELECT  nome_navio, classificacao_risco, indice_conformidade 
FROM tb_navios 
WHERE classificacao_risco = 'A' AND indice_conformidade >= 95;
```

## Quais embarcações têm classificação de risco C ou D e índice de conformidade menor ou igual a 95%?

```python
SELECT  nome_navio, classificacao_risco, indice_conformidade 
FROM tb_navios 
WHERE classificacao_risco IN ('C', 'D') 
AND indice_conformidade <= 95
```

## Quais embarcações têm classificação de risco A ou pontuação de risco igual a 0?

```python
SELECT  nome_navio, classificacao_risco, pontuacao_risco 
FROM tb_navios 
WHERE classificacao_risco = 'A' 
OR pontuacao_risco = 0;
```

## Quais embarcações foram inspecionadas em Dezembro de 2016?

```python
SELECT  nome_navio,temporada 
FROM tb_navios 
WHERE temporada LIKE '%DEZEMBRO 2016%' 

```
