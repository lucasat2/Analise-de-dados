```python
# Exercício 1 - Crie um objeto a partir da classe abaixo, chamado roc1, 
# passando 2 parâmetros e depois faça uma chamada aos atributos e métodos
from math import sqrt

class Rocket():
     
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
        
    def move_rocket(self, x_increment=0, y_increment=1):
        self.x += x_increment
        self.y += y_increment
        
    def print_rocket(self):
        print(self.x, self.y)
        
```

```python
roc1 = Rocket(50,24)
```

```python
roc1.print_rocket()
```

```python
roc1.move_rocket(17,18)
```

```python
roc1.print_rocket()
```

```python
# Exercício 3 - Crie a classe Smartphone com 2 atributos, tamanho e interface e crie a classe MP3Player com os 
# atributos capacidade. A classe MP3player deve herdar os atributos da classe Smartphone.
class Smartphone:
    
    def __init__(self,tamanho,interface):
        self.tamanho = tamanho
        self.interface = interface
  
```

```python

class MP3Player(Smartphone):
    def __init__(self, capacidade, tamanho = "Normal", interface = "Led"):
        self.capacidade = capacidade
        Smartphone.__init__(self, tamanho, interface)
       

    def imprimirMP3(self):
        print("Os valores para o equipamento são: %s %s %s" %(self.tamanho, self.interface, self.capacidade))
        
```

```python
aparelho = MP3Player('64 GB')
```

```python
aparelho.imprimirMP3()
```
