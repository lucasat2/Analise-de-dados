### GROUP BY ### 

Imagine que sua tabela de vendas cresceu e agora você tem vários registros do mesmo item. O seu chefe quer saber: "Quanto a gente faturou no total por cada tipo de item?".

"Separe todos os itens iguais em montinhos e some os valores de cada montinho".

```python
resumo_vendas = vendas.groupby('item')['preco_total'].sum()
print(resumo_vendas)
```

## MERGE

O merge é como passar uma supercola em duas tabelas diferentes e juntá-las usando uma "peça de encaixe" em comum (como um código de barras ou o nome do item).

```python
import pandas as pd

# 1. O sistema do estoque mandou esta tabela nova:
dados_estoque = [
    {'item': 'pneu', 'categoria': 'Borracharia'},
    {'item': 'oleo', 'categoria': 'Lubrificantes'},
    {'item': 'filtro', 'categoria': 'Peças'}
]
tabela_estoque = pd.DataFrame(dados_estoque)

# 2. A mágica da união (colando as duas):
tabela_completa = pd.merge(vendas, tabela_estoque, on='item')

# Mostra o resultado final
tabela_completa.head()
```

## Módulo 2: O Teste de Carga Pesada - TITANIC.

```python
import pandas as pd

# Puxando um arquivo real diretamente de um link da internet
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
dados_pesados = pd.read_csv(url)

# Tirando as medidas do arquivo
print(dados_pesados.shape)
```

O .shape (forma) funciona como uma trena de pedreiro. Ele mede a sua tabela e te devolve dois números: primeiro a quantidade exata de linhas, depois a de colunas. Isso serve para você saber o tamanho do "terreno" que vai capinar antes de começar a trabalhar.

```python
dados_pesados.head()
```

```python
dados_pesados.isnull().sum()

#isnull() pergunta "está vazio?" e o sum() soma todas as vezes que a resposta foi "sim".
```

```python
# Excluindo a coluna 'Cabin' inteira (axis=1 indica coluna)
dados_limpos = dados_pesados.drop('Cabin', axis=1)

# Preenchendo os vazios de 'Age' com a média exata das idades
media_idade = dados_limpos['Age'].mean()
dados_limpos['Age'] = dados_limpos['Age'].fillna(media_idade)

# Excluindo as linhas vazias restantes (dropna padrão age nas linhas)
dados_limpos = dados_limpos.dropna()

# Verificando o resultado da faxina
print(dados_limpos.isnull().sum())
```

mean() (A Média): O Python olhou para todo mundo que preencheu a idade, somou tudo e tirou a média. É como descobrir que a idade média do pessoal da festa é de 29 anos.

fillna() (A Massa Corrida): Esse comando pegou um carimbo com o número "29" e saiu preenchendo apenas os espaços que estavam em branco.

### Sobreviventes do titanic 

Na coluna Survived (Sobreviveu), o número 0 representa quem não sobreviveu e o 1 representa quem sobreviveu. Ao calcularmos a média matemática dessa coluna, o Python automaticamente nos entrega a porcentagem de sobreviventes.

```python
# Agrupando por gênero e calculando a média de sobrevivência (multiplicado por 100 para virar porcentagem)
taxa_sobreviventes = dados_limpos.groupby('Sex')['Survived'].mean() * 100

print("Taxa de sobrevivência por gênero (%):")
print(taxa_sobreviventes)
```

```python

# Escolhendo cores amigáveis (rosa e azul)
cores = ['#ff9999', '#66b3ff']

# Criando o gráfico (figsize define a largura e altura da imagem)
taxa_sobreviventes.plot(kind='bar', color=cores, figsize=(8, 5))

# Colocando os títulos e eixos
plt.title('Taxa de Sobrevivência por Gênero - Titanic', fontsize=14)
plt.ylabel('Sobrevivência (%)')
plt.xlabel('Gênero')

# xticks(rotation=0) deixa as palavras 'female' e 'male' na horizontal (retas)
plt.xticks(rotation=0)

# Mostra o resultado final
plt.show()
```

Lembra do nosso "segurança de balada" (o filtro com colchetes)? Vamos usá-lo para separar apenas os passageiros que pagaram mais caro (Pclass == 1) e, logo em seguida, descobrir quantos deles se salvaram.

```python
# 1. O segurança separando apenas a 1ª classe
primeira_classe = dados_limpos[dados_limpos['Pclass'] == 1]

# 2. Calculando a porcentagem de sobrevivência só desse grupo
taxa_vip = primeira_classe['Survived'].mean() * 100

print(f"Sobrevivência na 1ª Classe: {taxa_vip:.1f}%")
```

## Tabela Dinâmica (Pivot Table).

```python
# Cruzando Classe, Gênero e Taxa de Sobrevivência
tabela_dinamica = dados_limpos.pivot_table(values='Survived', index='Pclass', columns='Sex', aggfunc='mean') * 100

print(tabela_dinamica)
```

```python
tabela_dinamica.to_csv('resumo_titanic.csv')
```

### SEABORN - GRAFICOS MAIS BONITOS

Vamos começar criando um gráfico de distribuição. Pense nele como o perfil de uma montanha: a parte mais alta (o pico) mostra onde está a maior aglomeração de dados. Vamos usar isso para ver as idades dos passageiros.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Criando um histograma com a linha de tendência (a curva suave chamada KDE)
sns.histplot(dados_limpos['Age'], kde=True, color='purple')

plt.title('Montanha de Idades do Titanic')
plt.xlabel('Idade')
plt.ylabel('Quantidade de Pessoas')

plt.show()
```

### Vendo quem sobreviveu e nao sobreviveu por sexo 

O parâmetro hue='Sex' é o grande truque aqui: ele pega a contagem total e a divide em duas barras paralelas, uma para cada gênero.

```python
# O countplot conta a quantidade de pessoas, e o 'hue' divide pelas cores do gênero
sns.countplot(data=dados_limpos, x='Survived', hue='Sex', palette='Set2')

plt.title('Sobrevivência por Gênero')
# Trocando os números 0 e 1 por palavras para facilitar a leitura
plt.xticks([0, 1], ['Não Sobreviveu', 'Sobreviveu'])
plt.xlabel('')
plt.ylabel('Quantidade de Pessoas')

plt.show()
```

### Boxplot (Gráfico de Caixa)

Ele funciona como uma "radiografia": mostra onde a maior parte dos dados se concentra e também aponta os pontos fora da curva (quem foge muito do padrão).

Vamos tirar uma radiografia da idade dos passageiros, separando pelas três classes do navio

Caixa e as Linhas: Representam a "zona de normalidade" (a regra).

As Bolinhas: São os dados que "escaparam" dessa zona (as exceções). Uma bolinha isolada é uma exceção solitária. Um aglomerado de bolinhas são várias exceções juntas (como vários idosos viajando na 3ª classe, o que fugia da regra daquele grupo).

```python
sns.boxplot(data=dados_limpos, x='Pclass', y='Age', hue='Pclass', palette='Set3', legend=False)

plt.title('Radiografia de Idades por Classe')
plt.xlabel('Classe')
plt.ylabel('Idade')

plt.show()
```

A caixa verde da 1ª classe é a que fica mais no alto.

Isso faz todo sentido na vida real: como as passagens VIP eram caríssimas, elas atraíam pessoas mais velhas e com a vida financeira já estabilizada. Já a 3ª classe (a caixa mais baixa) era o "ônibus lotado" de jovens buscando uma vida nova.

### MACHINE LEARNING

```python
import pandas as pd

# Trabalhando com textos (e-mails) em vez de números
dados_email = {
    'mensagem': ['Ganhe dinheiro rapido', 'Reuniao de equipe amanha', 'Promocao exclusiva clique', 'Relatorio mensal anexado'],
    'tipo': ['Spam', 'Trabalho', 'Spam', 'Trabalho']
}

df_texto = pd.DataFrame(dados_email)
print(df_texto)
```

LIQUIDIFICADOR DE PALAVRAS PARA NUMEROS 

Aqui está o nosso "liquidificador" em ação. No Python, a ferramenta que faz isso se chama CountVectorizer.

Ele pega todas as frases, separa palavra por palavra e cria uma coluna para cada uma. Se a palavra aparece no e-mail, ele marca 1 (sim). Se não aparece, marca 0 (não).

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# 1. Ligando o liquidificador
liquidificador = CountVectorizer()

# 2. Jogando os textos dentro e batendo (transformando em números)
textos_triturados = liquidificador.fit_transform(df_texto['mensagem'])

# 3. Colocando o resultado em uma tabela nova para a gente conseguir ver
tabela_numeros = pd.DataFrame(textos_triturados.toarray(), columns=liquidificador.get_feature_names_out())

print(tabela_numeros)
```

```python
# X = as pistas (os textos triturados em números)
X = textos_triturados

# y = o gabarito (a coluna com as categorias 'Spam' ou 'Trabalho')
y = df_texto['tipo']
```

```python
from sklearn.tree import DecisionTreeClassifier

modelo_spam = DecisionTreeClassifier()
```

Agora vamos entregar o material de estudo para o nosso robô. É como dar a ele uma pilha de provas antigas (nossas pistas X) com as notas já corrigidas (nosso gabarito y) para ele descobrir sozinho a regra do jogo.

```python
# O comando '.fit' é o botão de "Estudar e Aprender"
modelo_spam.fit(X, y)

print("Pronto! O robô estudou e aprendeu o padrão.")
```

```python
# 1. Chegou um e-mail novo na sua caixa de entrada
email_novo = ["clique aqui e ganhe dinheiro"]

# 2. Passando no liquidificador para virar números (usamos só 'transform' agora)
email_novo_numeros = liquidificador.transform(email_novo)

# 3. A bola de cristal: pedindo para o robô adivinhar
previsao = modelo_spam.predict(email_novo_numeros)

print("O robô classificou este e-mail como:", previsao[0])
```

***Agora que seu robô sabe adivinhar, precisamos dar uma Nota de Boletim para ele. Em Machine Learning, chamamos isso de Acurácia (a porcentagem de acertos).Mas tem um truque: não podemos testar o robô com as exatas mesmas perguntas que ele usou para estudar, senão ele vai tirar 10 só porque decorou as respostas! Precisamos separar e esconder uma parte dos dados originais para fazer uma "prova surpresa" depois***
