# INNER JOIN (INTERSEÇÃO EM CONJUNTOS)

```python
# QUERO SABER QUAIS NOMES DE CLIENTES FIZERAM UM PEDIDO

SELECT P.id_pedido, C.nome_cliente  #SELECIONE O NOME DAS COLUNAS QUE QUER NAS 2 TABELAS 
FROM TB_PEDIDOS as P  # DA PRIMEIRA TABELA
INNER JOIN TB_CLIENTES as C # JUNTE COM A SEGUNDA TABELA
ON P.id_cliente = C.id_cliente;  #LIGACAO DAS TABELAS
```

# INNER JOIN DE 3 TABELAS

```python
# RETORNAR ID DO PEDIDO, NOME DO CLIENTE E NOME DO VENDEDOR 

SELECT 
    P.id_pedido, 
    C.nome_cliente, 
    V.nome_vendedor
FROM TB_PEDIDOS AS P
INNER JOIN TB_CLIENTES AS C ON P.id_cliente = C.id_cliente
INNER JOIN TB_VENDEDOR AS V ON P.id_vendedor = V.id_vendedor;
```

# LEFT JOIN

```python
# RETORNAR TODOS OS CLIENTES COM OU SEM PEDIDOS ASSOCIADOS.
# QUEREMOS OS DADOS DA TABELA DA ESQUERDA MESMO SEM CORRESPONDENCIA NA TABELA DA DIREITA

SELECT C.nome_cliente,P.id_pedido  
FROM TB_CLIENTES as C  
LEFT JOIN TB_PEDIDOS as P 
ON P.id_cliente = C.id_cliente;  
```

# RIGHT JOIN 

```python
SELECT C.nome_cliente,P.id_pedido  
FROM TB_PEDIDOS as P  
RIGHT JOIN TB_CLIENTES as C 
ON P.id_cliente = C.id_cliente;  
```

# INTEGRIDADE REFERENCIAL - DETECTORA DE REGISTROS ORFAOS - FULL JOIN

***CRIANDO UM REGISTRO ORFAO PROPOSITALMENTE***
* Registro Órfão é um dado em uma tabela (ex: Pedido) que aponta para um registro que não existe mais na tabela principal
* Causa: Você deletou o "pai", mas esqueceu de apagar os "filhos".

```python
INSERT INTO join_schema.TB_PEDIDOS (id_pedido, id_cliente, id_vendedor, data_pedido, id_entrega)
VALUES (1004, 10, 2, NOW(), 25);
```

* O FULL JOIN nao funciona no mysql workbench para isso usaremos o UNION
* O UNION remove registros duplicados do LEFT E DO RIGHT, o que pode comprometer um pouco o desempenho

```python
SELECT C.nome_cliente, P.id_pedido
FROM TB_CLIENTES AS C
LEFT JOIN TB_PEDIDOS AS P USING(id_cliente)

UNION

SELECT C.nome_cliente, P.id_pedido
FROM TB_CLIENTES AS C
RIGHT JOIN TB_PEDIDOS AS P USING(id_cliente);
```

* UNION ALL
* O Union All junta todos os resultados do LEFT E DO RIGHT empilhando tudo sem remover nada.

```python
SELECT C.nome_cliente, P.id_pedido
FROM TB_CLIENTES AS C
LEFT JOIN TB_PEDIDOS AS P USING(id_cliente)

UNION ALL

SELECT C.nome_cliente, P.id_pedido
FROM TB_CLIENTES AS C
RIGHT JOIN TB_PEDIDOS AS P USING(id_cliente);
```
