# GROUP BY

```python
# Contagem de clientes por cidade
# O agrupamento só pode ser feito pela coluna que nao tem calculo.

SELECT COUNT(id_cliente) AS contagem, cidade_cliente
FROM tb_clientes
GROUP BY cidade_cliente
ORDER BY contagem DESC;
```

# FUNCOES DE AGREGACAO

```python
# Quando usamos uma funcao de agregacao, nao é necessario usar group by, pois estamos apenas fazendo um calculo
```

* Media do valor dos pedidos

```python
SELECT AVG(valor_pedido) AS media
FROM TB_PEDIDOS;
```

* Media do valor dos pedidos por cidade, arredondando para 2 casas decimais

```python
SELECT ROUND(AVG(valor_pedido),2) AS media, cidade_cliente
FROM TB_PEDIDOS P
JOIN TB_CLIENTES C USING(id_cliente)
GROUP BY cidade_cliente
ORDER BY media DESC;
```

```python
SELECT 
    COALESCE(ROUND(AVG(valor_pedido), 2), 0) AS media,   # Retorna 0 ao achar um valor nulo
    cidade_cliente
FROM TB_PEDIDOS P
RIGHT JOIN TB_CLIENTES C USING(id_cliente)
GROUP BY cidade_cliente
ORDER BY media DESC;
```

# Soma(total) do valor dos pedidos por cidade e estado com right join para incluir cidades sem pedidos

```python

SELECT 
    C.cidade_cliente, 
    C.estado_cliente, 
    COALESCE(SUM(P.valor_pedido), 0) AS total   # Retorna 0 ao achar um valor nulo
FROM TB_PEDIDOS P
RIGHT JOIN TB_CLIENTES C USING(id_cliente)
GROUP BY C.cidade_cliente, C.estado_cliente
ORDER BY C.estado_cliente;
```

* Supondo que a comissão de cada vendedor seja de 10%, quanto cada vendedor ganhou de comissão nas vendas no estado do Ceará?
* Retorne 0 se não houve ganho de comissão

```python
SELECT V.nome_vendedor,

SUM(COALESCE(P.valor_pedido, 0)) * 0.10 AS comissao

FROM TB_VENDEDOR AS V

LEFT JOIN TB_PEDIDOS AS P USING (id_vendedor)

LEFT JOIN TB_CLIENTES AS C 

ON P.id_cliente = C.id_cliente 

AND C.estado_cliente = 'CE'

GROUP BY V.nome_vendedor;
```

# DISTINCT - Contar sem repeticao

# COUNT - NOME E CIDADE DO CLIENTE E NUMERO DE PEDIDOS

```python
SELECT 
    C.nome_cliente, 
    C.cidade_cliente, 
    COUNT(P.id_pedido) AS num_pedidos
FROM TB_PEDIDOS AS P
INNER JOIN TB_CLIENTES AS C USING (id_cliente)
WHERE 
    C.estado_cliente = 'CE'
GROUP BY 
    C.nome_cliente,cidade_cliente;
```

# Algum vendedo participou de vendas cujo pedido tenha sido superior a 600 no estado de Sp?

```python
SELECT 
    nome_cliente,cidade_cliente,nome_vendedor,valor_pedido
FROM TB_VENDEDOR AS V
INNER JOIN TB_PEDIDOS AS P USING (id_vendedor)
INNER JOIN TB_CLIENTES AS C USING (id_cliente)
WHERE 
    C.estado_cliente = 'SP' 
    AND P.valor_pedido > 600;

    --Nao houve obrigatoriedade de usar groupby aqui, foi apenas um filtro.
```

# HAVING - PARA FILTRAR UM GROUP BY

```python

-- Algum vendedor participou de vendas em que a média do valor_pedido por estado do cliente 
-- foi superior a 800?

SELECT 
    estado_cliente, nome_vendedor,
    CEILING(AVG(valor_pedido)) AS media_pedidos
    FROM TB_PEDIDOS P
    INNER JOIN TB_CLIENTES C USING (id_cliente)
    INNER JOIN TB_VENDEDOR V USING (id_vendedor)
    GROUP BY estado_cliente,nome_vendedor
    HAVING media_pedidos > 800 -- estamos filtrando dentro da media
    ORDER BY nome_vendedor;

```

```python
-- Qual estado teve mais de 5 pedidos?
SELECT 
    estado_cliente, COUNT(id_pedido) AS total_pedidos
FROM TB_PEDIDOS P
INNER JOIN TB_CLIENTES C USING (id_cliente)
GROUP BY estado_cliente
HAVING total_pedidos > 5
ORDER BY total_pedidos DESC;
```

# ROLLUP - Somar valor total do group by

```python
SELECT * FROM TB_VENDAS;
```

```python
SELECT ano,SUM(faturamento) AS total_faturamento
FROM TB_VENDAS
GROUP BY ano;
```

```python
SELECT ano,SUM(faturamento) AS total_faturamento
FROM TB_VENDAS
GROUP BY ano WITH ROLLUP;
```

```python
SELECT ano,SUM(faturamento) AS total_faturamento
FROM TB_VENDAS
GROUP BY ano WITH ROLLUP
HAVING total_faturamento > 13000;
```

```python
SELECT ano,SUM(faturamento) AS total_faturamento
FROM TB_VENDAS
GROUP BY ano WITH ROLLUP
ORDER BY ano;
```

```python
SELECT ano, pais,SUM(faturamento) AS total_faturamento
FROM TB_VENDAS
GROUP BY ano , pais with ROLLUP;
```

```python
SELECT ano, produto,SUM(faturamento) AS total_faturamento
FROM TB_VENDAS
GROUP BY ano , produto with ROLLUP;
```

# GROUPING - Subir os Totais

```python
SELECT ano, produto,SUM(faturamento) AS total_faturamento
FROM TB_VENDAS
GROUP BY ano , produto with ROLLUP
ORDER BY GROUPING(produto) DESC;
```

# ELIMINANDO OS NULOS DO ROLLUP
* Esses valores nulos gerados pelo rollup não existem no banco de dados, por isso não podemos usar case , para isso usaremos grouping

```python
SELECT 
    IF(GROUPING(ano), 'Total dos anos', ano) AS ano,
    IF(GROUPING(pais), 'Total dos países', pais) AS pais,
    IF(GROUPING(produto), 'Total dos produtos', produto) AS produto,
    SUM(faturamento) AS total_faturamento
FROM TB_VENDAS
GROUP BY ano, pais, produto WITH ROLLUP;
```
